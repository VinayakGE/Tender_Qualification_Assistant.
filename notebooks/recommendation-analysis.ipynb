# Recommendation Engine Analysis

This notebook analyzes the accuracy and behavior of the recommendation engine
against actual bid outcomes recorded in the decision ledger.

## Purpose

- Measure prediction accuracy: does BID → bid submitted? does NO_BID → bid skipped or bid lost?
- Identify cases where system recommendation differed from human decision
- Analyze confidence score reliability (high confidence → correct more often?)
- Find bottleneck patterns: which bottlenecks appear most frequently?

## Setup

Requires entries in `decision-ledger.json` and at least some outcomes in `data/outcomes/outcomes-ledger.json`.

```python
import sys

sys.path.insert(0, "..")

from src.ledger.decisions import DecisionLedger
from src.ledger.outcomes import OutcomeTracker

ledger = DecisionLedger()
decisions = ledger.read_all()
print(f"Total decisions in ledger: {len(decisions)}")

tracker = OutcomeTracker()
summary = tracker.get_accuracy_summary()
print(f"Outcomes recorded: {summary.get('total_outcomes', 0)}")
print(f"Accuracy: {summary.get('accuracy', 'N/A')}")
```

## Bottleneck Frequency

```python
from collections import Counter

if decisions:
    bottlenecks = [d.get("primary_bottleneck") for d in decisions if d.get("primary_bottleneck")]
    bottleneck_counts = Counter(bottlenecks)
    print("Top Bottlenecks:")
    for bottleneck, count in bottleneck_counts.most_common(10):
        print(f"  {bottleneck}: {count}")
else:
    print("No decisions yet. Run the pipeline on some tenders first.")
```
